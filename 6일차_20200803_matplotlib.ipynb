```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots()  # Axes 만들기

#위와 같이 서브플롯 선언을해서 ax axes를 안만들어주면 밑의 ax. 대신 plt.로 입력해도 만들어짐
ax.plot(x, x, label='linear')  # linear라는 그래프 데이터 주기
ax.plot(x, x**2, label='quadratic')  # quardratic 그래프 데이터 주기
ax.plot(x, x**3, label='cubic')  # cubic데이터 주기
ax.set_xlabel('x label')  # axes에 x 라벨 붙여주기
ax.set_ylabel('y label')  # axes에 y 라벨 붙여주기
ax.set_title("Simple Plot")  # axes 제목 붙여주기
ax.legend()  # legend 붙여주기 (저기 짝대기 이름 써져 있는거)
```

```python
def my_plotter(ax, data1, data2, param_dict): #다음과 같은 함수를 만들기
    out = ax.plot(data1, data2, **param_dict)
    return out

data1, data2, data3, data4 = np.random.randn(4, 100) #랜덤 값을 주기
fig, ax = plt.subplots(1, 1) #그래프나올 자리 만들기
my_plotter(ax, data1, data2, {'marker': 'x'}) #함수에 랜덤값 넣어서 그래프 만들기
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```

```python
def my_plotter(ax, data1, data2, param_dict): #다음과 같은 함수를 만들기
    out = ax[0].plot(data1, **param_dict)
    out = ax[1].plot(data2, **param_dict)
    return out

data1, data2 = np.random.randn(2, 100)
fig, ax=plt.subplots(2)
my_plotter(ax, data1, data2, {'marker': 'x'})
```

```python
def my_plotter(ax, data1, data2, param_dict): #다음과 같은 함수를 만들기
    out = ax[0][0].plot(data1, **param_dict)
    out = ax[1][0].plot(data2, **param_dict)
    return out

data1, data2 = np.random.randn(2, 100)
fig, ax=plt.subplots(2,2)
my_plotter(ax, data1, data2, {'marker': 'x'})
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data, alpha=0.3) #뒤에 x값, y값, color값, scalar값?
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(10, 6))    #figure()안에 figsize를 통해서 바탕 크기 설정 가능

plt.subplot(131)  #figure 안에 subplot 1번부터 3번까지 중 1번에 그리겠다.
plt.bar(names, values)
plt.subplot(132)  #figure 안에 subplot 1번부터 3번까지 중 2번에 그리겠다.
plt.scatter(names, values)
plt.subplot(133)   #figure 안에 subplot 1번부터 3번까지 중 3번에 그리겠다.
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(50, .025, r'$\mu=100,\ \sigma=15$')  #특정한 위치에 문자 삽입하기
plt.axis([40, 160, 0, 0.03])
plt.grid(True)  #그리드 (격자선 표시)
plt.show()
```
