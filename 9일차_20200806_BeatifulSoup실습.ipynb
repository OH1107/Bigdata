# 관심사이트 수집

# 4개 이상 컬럼의 데이터 프레임 만들기

```python
from bs4 import BeautifulSoup as bs
import requests as re
url = "https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000000&listType=B10002"
html = re.get(url).text
```

```python
req = re.get(url)
req
```

```python
soup = bs(html, "html.parser")
soup
```

```python
all_ = soup.find("ul",{"class":"type_normal"})
all_
```

```python
pic = all_.find_all("div",{"class","thumb_area"})
pic
```

```python
# 사진 링크
df_img=[]
for item in pic:
    try:
        img = item.find("img")
        df_img.append(img["data-original"][0:-10])
        print("img = ", img["data-original"][0:-10])
    except:
        print("No image")
```

```python
img["data-original"][0:-10]
```

```python
name = all_.find_all("p",{"class","cont"})
name
```

```python
# 타이틀 제목
df_name=[]
for item in name:
    try:
        title = item.find("a")
        df_name.append(title["title"])
        print("Title = ", title["title"])
    except:
        print("None")
```

```python
p = all_.find_all("div",{"class","price"})
p
```

```python
p[0].text.replace("\n","")
```

```python
#가격
df_price=[]
b="모바일최저가격원"
for item in p:
    try:
        price = item.text.replace("\n","")
        for char in b:
            price=price.replace(char,"")
            
        df_price.append(price)
        print("Price = ",price)
    except:
        print("None")
```

```python
#상품평

view = all_.find("div",{"class","info"}).find("span",{"class":"mall"})
view
```

```python
view = all_.find_all("div",{"class":"info"})
view
```

```python
print(view[0].find("a",{"class","txt"}).find("em").text)
```

```python
df_view=[]
for item in view:
    try:
        v = item.find("a",{"class","txt"}).find("em").text[1:-1]
        df_view.append(v)
        print("View : ", v)
    except:
        print("None")
```

```python
v = item.find("a",{"class","txt"}).find("em").text[1:-1]
v
```

```python
import pandas as pd
df=pd.DataFrame({'Title':df_name, 'Price':df_price, 'Img_link':df_img, 'View':df_view})
```
