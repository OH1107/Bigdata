```python
from bs4 import BeautifulSoup as bs
import requests as re
```

```python
url = 'https://www.yna.co.kr/view/AKR20200204081000017?section=it/it'
response = re.get(url)
soup = bs(response.text,'html.parser')
```

```python
article = soup.find('div',{'class':"story-news article"})
article
```

```python
article_p = article.find_all('p')
article_p
```

```python
content = ''
for text in article_p:
    content += text.text
```

```python
print(content)
print(type(content))
```

# 텍스트 전처리

## 1.토큰화

```python
from konlpy.tag import Mecab
mecab = Mecab()
```

```python
mecab_token = mecab.morphs(content)
print(mecab_token)
```

## 2.품사 부착

```python
mecab_pos = mecab.pos(content)
print(mecab_pos)
```

## 3.불용어 처리

```python
from collections import Counter
Counter(mecab_pos).most_common()
```

```python
stopPos = ['SY','SC','SSO','SSC','SF', #기호 제외
          'JKS', 'JKC', 'JKG', 'JKO', 'JKB', 'JKV', 'JKQ', 'JC', 'JX', #조사 제외
          'EP', 'EF', 'EC', 'XSV+EP', 'XSV','XSV+EC',#어미 제외
           'MM','MAG','MAJ', #관형사 제외
           'VV','VX', 'XSV+ETM', 'XSA', 'VV+ETM', 'VV+ETN', 'VV+EC','VV+EP' #동사 제외
          ]
stopWord = ['서울','연합뉴스','기자','홍지인',
            'ljungberg','yna','co','kr',
            '저작권자','c','무단', '전재', '재','배포','금지','송고',
           '2020','02','04','11','33',
           '있','는','수','인','이','이날','을','만','였','함']
word = []
for tag in mecab_pos:
  if tag[1] not in stopPos:
    if tag[0] not in stopWord:
      word.append(tag [0])
print(word)
```
