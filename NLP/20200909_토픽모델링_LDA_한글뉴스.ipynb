## 데이터 호출, 라이브러리 호출

```python
import pandas as pd
import re 
from konlpy.tag import Mecab
m = Mecab()
```

```python
df = pd.read_csv('LDA_example.csv')
df.head()
```

## 전처리 전 raw 데이터

```python
content = pd.Series(df['content'])
content
```

## 전처리 후 토큰화된 데이터

```python
mecab = Mecab()

stop_Pos = ['JKS','JKC','JKG','JKO','JKB','JKV','JKQ','JC','JX',
           'SF','SE','SS','SP','SSO','SSC','SC','SY','SH','SL','SN',
           'MM','MAG','MAJ']

stop_word = ['께서','에서','이다','에게','으로','이랑','까지','부터','하다']

new_content = pd.Series(range(len(df)))

for i, text in enumerate(content):
    text = mecab.pos(text)
    text = [j for j in text if len(j[0])>1]
    text = [j for j in text if j[1] not in stop_Pos]
    text = [j[0] for j in text if j[0] not in stop_word]
    new_content[i] = ' '.join([j for j in text])
    
new_content
```

## LDA 시각화

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation

tfidf_vect = TfidfVectorizer()
tfidf = tfidf_vect.fit_transform(new_content)
lda = LatentDirichletAllocation(n_components=20, max_iter=20, learning_method='online')

lda_output = lda.fit_transform(tfidf)
```

```python
import pyLDAvis
import pyLDAvis.sklearn

pyLDAvis.enable_notebook()
vis = pyLDAvis.sklearn.prepare(lda, tfidf, tfidf_vect, mds='tsne')
pyLDAvis.display(vis)
```
