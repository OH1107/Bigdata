```python
text = open('c://Ohhhhh/WordCloud/alice.txt').read()
```

```python
# PIL - Python Image Libary : 이미지 파일을 다루는 라이브러리

import numpy as np
from PIL import Image
from wordcloud import STOPWORDS
```

```python
# 이미지 파일을 불러오기
# 이미지의 픽셀 데이터를 numpy로 저장하기를 동시에 수행

alice_mask = np.array(Image.open("c://Ohhhhh/WordCloud/alice_mask.png"))
```

```python
# STOPWORDS로 제외할 단어를 추가 할수 있음 (보통 기본적을 많이 쓰는 단어를 제외)

stopwords = set(STOPWORDS)

# said 란 단어를 제외목록에 추가
stopwords.add("said")
```

```python
alice_mask
```

```python
plt.figure(figsize=(8,8))
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation = "bilinear")
plt.axis("off")
```

```python
wc = WordCloud(background_color = 'white',
              max_words=2000,
              mask=alice_mask,
              stopwords=stopwords)
wc=wc.generate(text)
wc
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc,interpolation = 'bilinear')
plt.axis("off")
```
