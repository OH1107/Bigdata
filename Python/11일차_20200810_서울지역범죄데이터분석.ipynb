### 소계(발생), 소계(검거) 컬럼 추가 ( 각 발생 검거 항목의 합계)

```python
# 절도 발생과 폭력 값이 숫자가 아니기에 더할수가 없음.
# 데이터 형태를바꿔줘야함

# 문자열 내의 ,를 제거
for tuple in df.values:
    for ele in tuple[-1:-4:-1]:
        str(ele).replace(',','')   # 아직은 데이터 수정된 사항을 리턴값 지정하지 않았음

# 우리까 바꿀 대상인데이터프레임 슬라이싱
textData = df[df.columns[-4:]].values
textData
```

```python
# ,가 제거된 문자열화된 숫자를 int형으로 바꾸기

for idx, row in enumerate(textData):
    for idx2, ele in enumerate(row):
        textData[idx][idx2] = int(str(ele).replace(',',''))
        
textData
df[df.columns[-4:]] = textData
df.head()
```

```python
df.values
```

```python
df.info()
```

```python
df['소계(발생)']= df['살인(발생)']+df['강도(발생)']+\
    df['강간(발생)']+df['절도(발생)']+df['폭력(발생)']
df['소계(검거)']= df['살인(검거)']+df['강도(검거)']+\
    df['강간(검거)']+df['절도(검거)']+df['폭력(검거)']
df
```

## 경찰서가 위치한 '구'에 대한 매핑

```python
policeToArea = {'서대문서':'서대문구','수서서':'강남구','강서서':'강서구','서초서':'서초구',
               '서부서':'은평구','중부서':'중구','종로서':'종로구','남대문서':'중구',
               '혜화서':'종로구','용산서':'용산구','성북서':'성복구','동대문서':'동대문구',
                '마포서':'마포구','영등포서':'영등포구','성동서':'성동구','동작서':'동작구',
               '광진서':'광진구','강북서':'강북구','금천서':'금천구','중랑서':'중랑구',
               '강남서':'강남구','관악서':'관악구','강동서':'강동구','종암서':'성북구',
               '구로서':'구로구','양천서':'양천구','송파서':'송파구','노원서':'노원구',
               '방배서':'서초구','은평서':'은평구','도봉서':'도봉구'}
df['구별'] = df['관서명'].apply(lambda v:policeToArea.get(v,v))
df.sort_values(by='구별').head()
```

```python
df_by_gu = pd.pivot_table(df, index='구별', aggfunc=np.sum)
df_by_gu.head()
```

```python
df_by_gu['강간검거율'] = df_by_gu['강간(검거)']/df_by_gu['강간(발생)']*100
df_by_gu['강도검거율'] = df_by_gu['강도(검거)']/df_by_gu['강도(발생)']*100
df_by_gu['살인검거율'] = df_by_gu['살인(검거)']/df_by_gu['살인(발생)']*100
df_by_gu['절도검거율'] = df_by_gu['절도(검거)']/df_by_gu['절도(발생)']*100
df_by_gu['폭력검거율'] = df_by_gu['폭력(검거)']/df_by_gu['폭력(발생)']*100
df_by_gu['검거율'] = df_by_gu['소계(검거)']/df_by_gu['소계(발생)']*100

del df_by_gu['강간(검거)']
del df_by_gu['강도(검거)']
del df_by_gu['살인(검거)']
del df_by_gu['절도(검거)']
del df_by_gu['폭력(검거)']

df_by_gu.head()
```

```python
#위 dataframe에서 검거율이 100% 를 초과한 값이 존재하는데 이는 작년에 발생한 범죄에 대해 다음해에 검거하게되는 경우 발생할 수 있는 값이다.
#100%이상인 값을 100%로 고정시켜 데이터를 다듬든다.

df_by_gu[df_by_gu[['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']] > 100] = 100
df_by_gu.head(10)
```

```python
#데이터 분석에 필요한 속성만을 얻고 ( feature reduction ) 속성의 이름을 재설정한다.

df_by_gu.rename(columns = {'강간(발생)':'강간', 
                       '강도(발생)':'강도', 
                       '살인(발생)':'살인', 
                       '절도(발생)':'절도', 
                       '폭력(발생)':'폭력'}, inplace=True)
#del df_by_gu['소계(발생)']
del df_by_gu['소계(검거)']

df_by_gu.head()
```

```python
# 위 dataset에 구별 '인구수' 속성을 추가시키기 위해서 seoul_population.csv dataset을 통해 각 구별 인구수에 대한 데이터를 추가시킨다. 
# 두 dataset간 index가 '구'의 이름으로 일치한다.
```

# Data Visualization
### 각 구별로 범죄상황에 대해 시각화
### ['강간','강도','살인','절도','폭력'] 속성에 대해서는 각 값이 절대량을 나타내는 값이므로 시각화를 위해 각 속성별로 Normailization 하여 [0,1]로 값을 변경한다. ( data transformation )

## Normailization

##### 절대값 데이터 수치를 비교하기에는 수차이가 너무 나기에 표준화를 해야함
##### (데이터값 - 최소값 )/ 최대값   - > 1보다는 무조건 큰값이 나오게 되어있음

```python
target_col = ['강간', '강도', '살인', '절도', '폭력','소계(발생)']
max_column = df_by_gu[target_col].max() # vector
min_column = df_by_gu[target_col].min()
crime_count_norm = (df_by_gu[target_col] - min_column)/max_column
crime_count_norm.sort_values(by='소계(발생)',ascending=False).head()
```

## Data Visualization 1 : 서울시 각 구별 범죄발생률

```python
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

plt.figure(figsize = (10,10))
sns.heatmap(crime_count_norm.sort_values(by='소계(발생)', ascending=False), annot=True, fmt='f', linewidths=.5)
plt.title('서울시 구별 강력범죄 발생빈도')
plt.show()
```

## Data Visualization 2 : 서울시 구별 인구수 대비 범죄율

```python
target_col = ['강간', '강도', '살인', '절도', '폭력','소계(발생)']
frame_temp = df_by_gu[target_col].div(df_by_gu['인구수'],axis=0)
max_column = frame_temp.max() # vector
min_column = frame_temp.min()
crime_count_norm = (frame_temp - min_column)/max_column
crime_count_norm.head()
```

```python

plt.figure(figsize = (10,10))
sns.heatmap(crime_count_norm.sort_values(by='소계(발생)', ascending=False), annot=True, fmt='f', linewidths=.5)
plt.title('서울시 구별 인구수대비 강력범죄 발생빈도')
plt.show()
```

## Data Visualization 3 : 각 구별 살인 횟수

```python
import json
import folium
import warnings
warnings.simplefilter(action = "ignore", category = FutureWarning)

geo_path = 'c://Ohhhhh/seoul_crime_data/skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=11, tiles='Stamen Toner')

map.choropleth(geo_data = geo_str,
               data = df_by_gu['살인'],
               columns = [df_by_gu.index, df_by_gu['살인']],
               fill_color = 'PuRd', #PuRd, YlGnBu
               key_on = 'feature.id')
map
```
