# 이름에서 특정 문자열 추출하기

# MR, Miss, Mrs, Master 등 확인 가능한 정보들 있음

# 해당 정보들을 정규표현식을 사용해서 추출하기

```python
# str.extrack (추출할 조건 내용을 입력하여 데이터에서 원하는 문자열을 추출)
# expand가 True 인 경우 출력을 하나의 column단위로 나타내고, 
# False인 경우 label단위 (시리즈인가?)로 출력

titanic['Title']=titanic['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
titanic['Title']
```

```python
titanic['Title'].value_counts()
```

# 추출한 데이터를 숫자화하여 수정하기

#### mr=0, miss=1, mrs=2
#### 나머지 3

```python
title1=[]
for i in titanic['Title']:
    if (i=='Mr'):
        title1.append(0)
        
    elif (i=='Miss'):
        title1.append(1)
    
    elif (i=='Mrs'):
        title1.append(2)
        
    else:
        title1.append(3)
        
titanic["Title1"]=title1
titanic["Title1"]
```


#### 다른 방법 titanic["Title1"]을 만듦 (13번째 칼럼)
#### titanic.iloc[0,12] ('Title' 칼럼)가 mr인지 비교하고 titanic.iloc[0,13]에 0값을 추가
#### 위의 동작을 for 반복문으로 반복



```python
titanic["Title1"]=9

for row in range(0, len(titanic)):
    if titanic.iloc[row, 12]=="Mr":
        titanic.iloc[row, 13]=0
    elif titanic.iloc[row, 12]=="Miss":
        titanic.iloc[row, 13]=1
    elif titanic.iloc[row, 12]=="Mrs":
        titanic.iloc[row, 13]=2
    else:
        titanic.iloc[row, 13]=3
```

```python
titanic.head(10)
```

```python
# map 메소드를 사용해서 지정된 값을 원하는 값으로 변환 할 수 있음 (튜플 리스트 딕셔너리 다 가능한듯)

title_mapping = {"Mr":0, "Miss":1,"Mrs":2,"Master":3,"Dr":3,"Rev":3,"Col":3, "Mlle":3, "Major":3,"Sir":3, "Countess":3, "Jonkheer":3,
                "Don":3, "Lady":3, "Ms":3, "Capt":3, "Mme":3}

titanic["Title2"]=titanic["Title"].map(title_mapping)
```

```python
# 나이 구간 확인을 용이하게 하기 위해 함수만들기
# x축의 최대최소값, 칼럼을 입력하면 나오는 그래프 함수 만들기

def chart_kdeplot (feature, x_min, x_max):
    facet=sns.FacetGrid(titanic, hue="Survived",aspect=4)

    facet.map(sns.kdeplot,feature,shade=True)
    facet.set(xlim=(0,titanic[feature].max()))
    facet.add_legend()
    plt.xlim(x_min,x_max)
```

```python
chart_kdeplot('Age',20,40)
```

#### 나이를 카테고리화 시킬거임
#### child: 0
#### young :1
#### adult : 2
#### mid-age : 3
#### senior : 4

```python
# 데이터 값을 정형화하는 다른 방법임
# 기존에 알던 인덱싱 loc와는 다른 기능을 함

# df.loc[데이터의 조건, 적용시킬 컬럼(신규도 가능)] = 적용할 값

titanic.loc[titanic['Age']<=16, 'Age']=0,
titanic.loc[(titanic['Age']>16) & (titanic['Age']<=26), 'Age']=1,
titanic.loc[(titanic['Age']>26) & (titanic['Age']<=36), 'Age']=2,
titanic.loc[(titanic['Age']>36) & (titanic['Age']<=62), 'Age']=3,
titanic.loc[titanic['Age']>62, 'Age']=4
```

# Embarked 랭크 정량화하기
#### 승선한 항구명 (s, c, q로 구성되어있음)
#### s=0 c=1 q=2로 바꿔서 새로운 칼럼 만들기

```python
# 결측치 확인

titanic.isnull().sum()
```

```python
#승선권 (Pclass)별 Embarked 비율 확인하기 (결측치를 근사치로 매우기 위해서)

Pclass1=titanic[titanic['Pclass']==1]['Embarked'].value_counts()
Pclass2=titanic[titanic['Pclass']==2]['Embarked'].value_counts()
Pclass3=titanic[titanic['Pclass']==3]['Embarked'].value_counts()

df = pd.DataFrame([Pclass1, Pclass2, Pclass3])
df.index=['1st class','2nd class','3rd class']
df.plot(kind='bar', stacked=True, figsize=(5,5))
```

```python
# Embarked 결측치 매우기
titanic['Embarked1']=titanic['Embarked'].fillna('S')
titanic['Embarked1'].isnull().sum()
```

```python
# S : 0,     C: 1 ,   Q:2로 바꾸기 매핑으로

embarked_mapping = {'S':0, 'C':1, 'Q':2}

titanic['Embarked2']=titanic['Embarked1'].map(embarked_mapping)
```

# Family Size 정량화 하기
### SibSp : 동반한 형제 자매 수
