```python
#파라메터를 딕셔너리값으로 만들어주고 
# 인증키를 디코드해서 url 소스를 출력
# 단, 파라미터들을 딕셔너리 형태로 변수에 대입해놓으면
# 다음번에 호출하고 새로운 api를 만들때 용이

import requests

t_key_decode = requests.utils.unquote(t_key)

print(" my_key :", t_key)
print(" my_key_decode : ", t_key_decode)

# req_params에 서비스키부터 타입까지 입력

req_params = {'serviceKey' : t_key_decode,
             'LAWD_CD' : lawd_cd,
              'DEAL_YMD':deal_ymd,
             'type':"json"}

# 다음과 같은 함수 .get 은( 기본 url과 뒤의 파라메타 부분을 요구함/ 파라메타 부분은 딕셔너리형태로 대입해야함)
r = requests.get(base_url, params=req_params)
print(r.url)
```

```python
# url을 str형태로 받아오기

import urllib.request
res = urllib.request.urlopen(r.url)
res
```

```python
# 받아온 str 데이터를 decode('utf-8')로 한글 안깨지게 볼수 있음

# res 변수에 넣어놓은 url 데이터는 한번 서비스하고 날라가버림
# 그렇기때문에 컴파일을 처음부터 다시 돌려줘야함
# 아니면 위의 urllib.request 컴파일 칸에 붙여야함!

xml_str = res.read().decode('utf-8')
xml_str
```

```python
# xml파일을 json파일로 바꾸는 과정


import xmltodict
import json

# xmltodict 라이브러리를 사용해 xml_str을
# json 라이브러리의 json.dumps함수를 사용해 
# json_str으로 새로 변환

json_str=json.dumps(xmltodict.parse(xml_str), indent=4, ensure_ascii=False)

# json 형태로 위의 데이터를 보기 좋게 출력

json_object=json.loads(json_str) # json 형태로 위의 데이터를 재출력
json_object
```

```python
#판다스 모듈 중 json파일을 읽을 수 있는 라이브러리 호출

import pandas as pd
from pandas.io.json import json_normalize

# 판다스를 통해서 json_object에 저장되어 있는 데이터를 body중 item만 뽑아올 것임

json_object['response']['body']
```

```python
# 판다스 함수인 json_normalized을 이용해서 json파일을 판다스 데이터프레임 형태로 출력

# 내가 어떤 데이터를 뽑아낼지 확실하게 지시해줘야 함
df=pd.json_normalize(json_object['response']['body']['items']['item'])
df
```
