```python
import requests as re
from bs4 import BeautifulSoup as bs
url = 'https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001'
html = re.get(url).text
soup =bs(html,"html.parser")
```

```python
# 중간 값 확인하기
soup
```

```python
all_ = soup.find("ul",{"class":"type06_headline"})
all_
```

```python
dl = all_.find_all("dl")
dl
```

```python
for item in dl:
    try:
        img= item.find("dt",{"class":"photo"}).find("img")
        print("img = ",img["src"])
        print("-"*20)
    except:
        print("No image")
```

```python
for item in dl:
    link = item.find("dt",{"class":""}).find("a")
    print(link["href"])
    print(link.text.replace("\t","").replace("\n","").lstrip())
    print('-'*20)
```

```python
for item in dl:
    try:
        content = item.find("dd")
        print(content.text.replace("\n","").split("…")[0])
        print(content.text.replace("\n","").split("…")[1])
        print("\n")
    except:
        print("No Content")
```
